Random output data table with limit 5

```python
%%bigquery --project data-viz-big-query
SELECT *
FROM `bigquery-public-data.iowa_liquor_sales.sales`
LIMIT 5
```

liquor sales data in 2019 based in Des Moines.

```python
%%bigquery --project data-viz-big-query
SELECT city as city, SUM(sale_dollars) as total_sales, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2019-01-01' AND '2019-12-31' AND city = "Des Moines"
GROUP BY city, date
ORDER By date
```

liquor sales data based in POLK county in July of 2021.

```python
%%bigquery --project data-viz-big-query
SELECT city as city, county as county, item_description as item, SUM(sale_dollars) as total_sales, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2021-07-01' AND '2021-08-01' AND county = "POLK"
GROUP BY city, county, item, date
ORDER BY date
LIMIT 10
```

total amount of liquor sold in July of 2021 with volume

```python
%%bigquery --project data-viz-big-query
SELECT county as county, SUM(sale_dollars) as total_sales_amount, SUM(volume_sold_gallons) as volume_sold_gallons
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2021-07-01' AND '2021-08-01' AND county = "POLK"
GROUP BY county
```

total amount of liquor sold in July of 2020 with volume

```python
%%bigquery --project data-viz-big-query

SELECT county as county, SUM(sale_dollars) as total_sales_amount, SUM(volume_sold_gallons) as volume_sold_gallons
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2020-07-01' AND '2020-08-01' AND county = "POLK"
GROUP BY county
```

we can conclude that more amount of liquor was sold in 2021 looking at two tables above.
